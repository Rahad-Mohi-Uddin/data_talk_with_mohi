```python
#import necessary library

import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#load three dataset to be cleaned 

all_deaths = pd.read_csv("Data Talk/Smoking/number-of-deaths-by-risk-factor.csv")
death_rate = pd.read_csv("Data Talk/Smoking/death-rate-smoking.csv")
affordability = pd.read_csv("Data Talk/Smoking/affordability-cigarettes.csv")
```

```python
#first let's explore all_deaths data

all_deaths.head()
```

```python
#extracts Bangladesh's data from the dataset all_deaths

all_deaths_bangladesh = all_deaths.loc[all_deaths['Entity']=='Bangladesh']
all_deaths_bangladesh.head()
```

```python
#delete entity and code column 

all_deaths_bangladesh_v02 = all_deaths_bangladesh.drop(['Entity','Code'],axis=1)
all_deaths_bangladesh_v02.head()
```

```python
#renaming the elements of header row

all_deaths_bangladesh_v03 = all_deaths_bangladesh_v02.set_axis(['Year','Air Pollution','High Blood Pressure','Diet High in Sodium','Diet Low in Whole Grain',
                                                               'Alcohol Use','Diet Low in Fruits','Unsafe Water Source','Secondhand Smoke','Low Birth Weight',
                                                               'Child Wasting','Unsafe Sex','Diet Low in Nuts & Seeds','Household Air Pollution','Diet Low in Vegetables',
                                                               'Low Physical Activity','Smoking','High Fasting Plasma Glucose','Air Pollution','High BMI','Unsafe Sanitation',
                                                               'No Access to Handwashing Facility','Drug Use','Low Bone Mineral Density','Vitamin A Deficiency',
                                                               'Child Stunting','Discontinued Breastfeeding','Non-Exclusive Breastfeeding','Iron Deficiency'],axis=1,inplace=False)

all_deaths_bangladesh_v03.head()
```

```python
#adding a new column for total death

all_deaths_bangladesh_v03['Total Death'] = all_deaths_bangladesh_v03.iloc[:,1:29].sum(axis=1)
all_deaths_bangladesh_v03.head()
```

```python
#now let's extract bangladesh data from death rate dataset 

death_rate_sa = death_rate.loc[death_rate['Entity'].isin(['Bangladesh','India','Pakistan'])].drop(['Code'], axis=1)
death_rate_sa.head()
```

```python
#now pivot the dataset and make the year column index 

death_rate_sa_v02 = death_rate_sa.pivot(index="Year",columns="Entity",values="Deaths Rate")
death_rate_sa_v02.head()
```

```python
#now let's explore bangladesh data from affordability dataset

affordability_sa = affordability.loc[affordability['Entity'].isin(['Bangladesh','India','Pakistan'])].drop(['Code'], axis=1)
affordability_sa.head()
```

```python
#now pivot the dataset and make the year column index 

affordability_sa_v02 = affordability_sa.pivot(index="Year",columns="Entity",values="Affordability")
affordability_sa_v02
```

```python
#now save all the file cleaned as our desire 

affordability_sa_v02.to_csv('affordability_compare.csv',header=True,index=True)
```

```python
death_rate_sa_v02.to_csv('death_rate_compare.csv',header=True,index=True)
```

```python
all_deaths_bangladesh_v03.to_csv('bangladesh_death_factor.csv',header=True,index=False)
```
